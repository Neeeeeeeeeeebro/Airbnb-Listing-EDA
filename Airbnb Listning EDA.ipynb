```python
plt.figure(figsize=(10, 6))
sns.boxplot(data=df, x='room_type', y='price')
plt.title('Relationship between Room Type and Price')
plt.xlabel('Room Type')
plt.ylabel('Price')
plt.xticks(rotation=45)
plt.show()
```

```python
plt.figure(figsize=(12, 10))
sns.barplot(data=df, x='neighbourhood', y='price',color='r')
plt.title('Price Distribution across Neighbourhoods')
plt.xlabel('Neighbourhood')
plt.ylabel('Price')
plt.xticks(rotation=90)
plt.grid()
plt.legend()
plt.show()
```

```python
plt.figure(figsize=(10, 8))
sns.scatterplot(data=df, x='number_of_reviews', y='neighbourhood',color='y')
plt.title('Relationship between Number of Reviews and Minimum Nights')
plt.xlabel('Number of Reviews')
plt.ylabel('neighbourhood')
plt.grid()
plt.show()
```

```python
## Impact of Recent Reviews on Room Availability ##
plt.figure(figsize=(10, 6))
sns.boxplot(data=df, x='number_of_reviews_ltm', y='availability_365',color='b')
plt.title('Impact of Recent Reviews on Room Availability')
plt.xlabel('Number of Recent Reviews')
plt.ylabel('Room Availability')
plt.xticks(rotation=45)
plt.grid()
plt.show()
```

```python
### كيف يتوزع عدد الإعلانات المختلفة في الأحياء المختلفة؟ هل هناك أحياء يغلب فيها نوع معين من الإعلانات على الآخر؟ ###
```

```python
neighbourhood_counts = df['neighbourhood'].value_counts()
print(neighbourhood_counts)
```

```python
room_type_neighbourhood = pd.crosstab(df['neighbourhood'], df['room_type'])
print(room_type_neighbourhood)
```

```python
neighbourhood_counts.plot(kind='bar', figsize=(10, 6),color='r')
plt.title('Distribution of Listings by Neighbourhood')
plt.xlabel('Neighbourhood')
plt.ylabel('Number of Listings')
plt.grid()
plt.legend()
plt.show()
```

```python
## تحليل العلاقة بين عدد الممتلكات والسعر ## 
plt.boxplot([df[df['calculated_host_listings_count'] == count]['price'] for count in df['calculated_host_listings_count'].unique()])
plt.xticks(range(1, len(df['calculated_host_listings_count'].unique()) + 1), df['calculated_host_listings_count'].unique())
plt.title('Relationship between Host Listings Count and Price')
plt.xlabel('Host Listings Count')
plt.ylabel('Price')
plt.show()
```

```python
### the relation between number_of_reviews_ltm and calculated_host_listings_count ##
plt.scatter(df['number_of_reviews_ltm'], df['calculated_host_listings_count'],edgecolors='k')
plt.title('Relationship between Host Reviews and New Booking Count')
plt.xlabel('Number of Reviews (Last 12 Months)')
plt.ylabel('New Booking Count (Last 12 Months)')
plt.grid()
plt.legend()
plt.show()
```

```python
## analysis days free in year ####
plt.hist(df['availability_365'], bins=20, edgecolor='k',label='Availability Days',color='purple')
plt.title('Distribution of Availability Days in the Year')
plt.xlabel('Number of Available Days')
plt.ylabel('Frequency')
plt.grid()
plt.legend()
plt.show()
```

```python
### Analysis of the relationship between the number of available days and the number of bookings per year ###

plt.scatter(df['availability_365'], df['number_of_reviews_ltm'])
plt.title('Relationship between Availability Days and New Booking Count')
plt.xlabel('Number of Available Days')
plt.ylabel('New Booking Count (Last 12 Months)')
plt.show()
```

```python
### Calculating average room rates for each room type ###
average_prices_by_room_type = df.groupby('room_type')['price'].mean()
print(average_prices_by_room_type)
```

```python
## def expensive and least ###
most_expensive_type = average_prices_by_room_type.idxmax()
least_expensive_type = average_prices_by_room_type.idxmin()
print(f"The most expensive room type is: {most_expensive_type}")
print(f"The least expensive room type is: {least_expensive_type}")
```

```python
## show data between expensive and least ##
average_prices_by_room_type.plot(kind='bar', color='skyblue')
plt.title('Average Room Prices by Room Type')
plt.xlabel('Room Type')
plt.ylabel('Average Price')
plt.xticks(rotation=45)
plt.grid()
plt.legend()
plt.show()
```

```python
the_relation_between_values = df.groupby('host_id')['number_of_reviews_ltm'].median()
print(the_relation_between_values)
```

```python
plt.figure(figsize=(8,6))
the_relation_between_values.plot(kind='hist',color='g')
plt.title('the_relation_between_values')
plt.xlabel('host_id')
plt.ylabel('number_of_reviews_ltm')
plt.grid()
plt.xticks(rotation=90)
plt.legend()
plt.show()
```
